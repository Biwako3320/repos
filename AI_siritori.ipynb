```python
import MeCab
import random
import jaconv

def select_word(text):
  mecab = MeCab.Tagger("-Oyomi")
  node=mecab.parseToNode(text)
  words=[]
  while node:
    features = node.feature.split(",")
    if len(features)>7 and (features[0]=='名詞' or features[0]=='形容詞' or features[0]=='動詞'):
      word=features[7]
      word_hira=jaconv.kata2hira(word)

      if word_hira and word_hira[0] != 'ー' and word_hira[-1] != 'ー' and word_hira[0] != 'ん' and word_hira[-1] != 'ん' and word_hira[0] != 'ゃ' and word_hira[-1] != 'ゃ' and word_hira[0] != 'ゅ' and word_hira[-1] != 'ゅ' and word_hira[0] != 'ょ' and word_hira[-1] != 'ょ' and word_hira[0] != 'っ' and word_hira[-1] != 'っ' and len(word_hira) > 1:
        words.append(word_hira)
    node=node.next
  return words

def get_word(current_word,words, used_words):
  last_char=current_word[-1]
  for word in words:
    if word[0]==last_char and word not in used_words:
      return word
  return None

def shiritori(words):
  used_words=[]
  current_word=random.choice(words)
  print("最初の言葉:",current_word)
  used_words.append(current_word)


  while True:
    user=input("ユーザーの番です。言葉を入力してください。ひらがなかカタカナで入力してね:")
    user_hira=jaconv.kata2hira(user)

    if user_hira in used_words:
      print("2回目なので別の言葉を入力してください")
      continue

    if user_hira[0]!=current_word[-1]:
      print("しりとりが続行できません。もう一度入力してください")
      continue

    if len(user_hira) == 1:
      print('1文字は使用できません。もう一度入力してください')
      continue

    used_words.append(user_hira)
    current_word=get_word(user_hira,words,used_words)
    if current_word is None:
      print("これ以上言葉が続かないためしりとりを終了します")
      break

    if user_hira[-1]=='ん' :
      print("「ん」がついたのでユーザーの負けです。")
      break

    used_words.append(current_word)
    print("プログラムの番です:",current_word)

    if current_word[-1]=='ん':
      print("「ん」がついたのでプログラムの負けです")
      break


if __name__=="__main__":
  with open("Neko.txt","r",encoding="utf-8")as file:
    text_data=file.read()
  word_list=select_word(text_data)

  shiritori(word_list)
```

返す単語をランダムにする。辞書登録してないからユーザーはやりたい放題。「ん」がつく単語をなるべく避ける。
