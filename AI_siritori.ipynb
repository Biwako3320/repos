```python
import MeCab
import random
import jaconv

def select_word(text):
  mecab = MeCab.Tagger("-Oyomi")
  node=mecab.parseToNode(text)
  words=[]
  while node:
    features = node.feature.split(",")
    if len(features)>7 and (features[0]=='' or features[0]=='' or features[0]==''):
      word=features[7]
      word_hira=jaconv.kata2hira(word)

      if word_hira and word_hira[0] != '' and word_hira[-1] != '' and word_hira[0] != '' and word_hira[-1] != '' and word_hira[0] != '' and word_hira[-1] != '' and word_hira[0] != '' and word_hira[-1] != '' and word_hira[0] != '' and word_hira[-1] != '' and word_hira[0] != '' and word_hira[-1] != '' and len(word_hira) > 1:
        words.append(word_hira)
    node=node.next
  return words

def get_word(current_word,words, used_words):
  last_char=current_word[-1]
  for word in words:
    if word[0]==last_char and word not in used_words:
      return word
  return None

def shiritori(words):
  used_words=[]
  current_word=random.choice(words)
  print(":",current_word)
  used_words.append(current_word)


  while True:
    user=input(":")
    user_hira=jaconv.kata2hira(user)

    if user_hira in used_words:
      print("2")
      continue

    if user_hira[0]!=current_word[-1]:
      print("")
      continue

    if len(user_hira) == 1:
      print('1')
      continue

    used_words.append(user_hira)
    current_word=get_word(user_hira,words,used_words)
    if current_word is None:
      print("")
      break

    if user_hira[-1]=='' :
      print("")
      break

    used_words.append(current_word)
    print(":",current_word)

    if current_word[-1]=='':
      print("")
      break


if __name__=="__main__":
  with open("Neko.txt","r",encoding="utf-8")as file:
    text_data=file.read()
  word_list=select_word(text_data)

  shiritori(word_list)
```
